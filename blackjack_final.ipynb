```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
import math
import re
from classes import simulate_round, LCG
```

```python
def spawn_deck():
    '''
    Generate an ordered deck of cards with the card type and its point value
    By default the aces are worth 11
    '''
    deck = [('Ace of Hearts', 11), ('Two of Hearts', 2), ('Three of Hearts', 3), ('Four of Hearts', 4), ('Five of Hearts', 5), ('Six of Hearts', 6),
        ('Seven of Hearts', 7), ('Eight of Hearts', 8), ('Nine of Hearts', 9), ('Ten of Hearts', 10), ('Jack of Hearts', 10), ('Queen of Hearts', 10), ('King of Hearts', 10),
        ('Ace of Spades', 11), ('Two of Spades', 2), ('Three of Spades', 3), ('Four of Spades', 4), ('Five of Spades', 5), ('Six of Spades', 6),
        ('Seven of Spades', 7), ('Eight of Spades', 8), ('Nine of Spades', 9), ('Ten of Spades', 10), ('Jack of Spades', 10), ('Queen of Spades', 10), ('King of Spades', 10),
        ('Ace of Clubs', 11), ('Two of Clubs', 2), ('Three of Clubs', 3), ('Four of Clubs', 4), ('Five of Clubs', 5), ('Six of Clubs', 6),
        ('Seven of Clubs', 7), ('Eight of Clubs', 8), ('Nine of Clubs', 9), ('Ten of Clubs', 10), ('Jack of Clubs', 10), ('Queen of Clubs', 10), ('King of Clubs', 10),
        ('Ace of Diamonds', 11), ('Two of Diamonds', 2), ('Three of Diamonds', 3), ('Four of Diamonds', 4), ('Five of Diamonds', 5), ('Six of Diamonds', 6),
        ('Seven of Diamonds', 7), ('Eight of Diamonds', 8), ('Nine of Diamonds', 9), ('Ten of Diamonds', 10), ('Jack of Diamonds', 10), ('Queen of Diamonds', 10), ('King of Diamonds', 10)]
    return deck

def shuffle_deck(deck, seed, deck_num=1):
    '''
    Pass in a deck of cards and a seed to recreate a shuffling. You can shuffled multiple decks together with deck_num.
    Start with the last card in the deck, i
    Use the LCG class to generate the next random and use mod to scale the PRN between 0 and i. 
    Update seed
    swap cards for positions i and scaled PRN j
    '''
    deck = deck*deck_num #number of decks
    n = len(deck)
    for i in range(n-1, 0, -1): #start with last index of card in deck
        lcg = LCG(seed) #LCG instance
        next_seed = lcg.next_seed() #get seed
        j = next_seed%(i+1) #card to be switched with i_th card
        seed = next_seed
        deck[i], deck[j] = deck[j], deck[i] #swap position i in deck with position j
    return deck


def deal_hand(shuffled_deck):
    '''
    Deal the first 4 cards in the shuffled deck to the player and dealer
    '''
    dealt_hands = {'player_hand': [], 'dealer_hand':[]}
    for i in range(2):
        dealt_hands['player_hand'].append(shuffled_deck.pop(0))
        dealt_hands['dealer_hand'].append(shuffled_deck.pop(0))
    return dealt_hands

def cumulative_average(numbers):
    '''
    This function takes in an array of results, ex. [-1,0,1], and returns an array of same length of average score
    '''
    total_sum = 0
    count = 0
    averages = []
    for num in numbers:
        total_sum += num
        count += 1
        average = total_sum / count
        averages.append(average)
    return averages 
 
def niceWatch_runIt(conditions_dict):
    seed = conditions_dict['seed']
    deck_num = conditions_dict['number of decks']
    trials = conditions_dict['trials']
    strategies = conditions_dict['strategies']
    hit_thresh = conditions_dict['when to stand']
    
    basic_counter = 0
    game_results = {}
    for h,hit_val in enumerate(hit_thresh):
        
        for s,strat in enumerate(strategies):
            if strat == 'strategy_base' and basic_counter == 1: #if basic strat and not yet seen basic strat
                continue
            elif strat == 'strategy_base': #if basic strat and already seen a basic strat iteration
                basic_counter = 1
            
            trial_results = []
            for t in range(trials):
                deck = shuffle_deck(spawn_deck(), seed, deck_num=deck_num) #create and shuffle a deck
                hands = deal_hand(deck) #deal hands
                player_hand = hands['player_hand'] 
                dealer_hand = hands['dealer_hand']
                round_instance = simulate_round(player_hand, dealer_hand, deck) #get simulation instance of round
                if strat == 'strategy_soft':
                    trial_results.append((t,round_instance.strategy_soft(hit_val))) #tuple of round # and result of strategy 
                elif strat == 'strategy_hard':
                    trial_results.append((t,round_instance.strategy_hard(hit_val))) #tuple of round # and result of strategy
                elif strat == 'strategy_base':
                    trial_results.append((t,round_instance.strategy_base()))
                seed = LCG(seed).next_seed() #update seed to next seed
            
            cum_results = cumulative_average([i[1] for i in trial_results]) #tuple of # of games played and cumulative average result
            cum_results = [(i,e) for i,e in enumerate(cum_results)]
            if strat == 'strategy_base':
                game_results[f"basic_strategy"] = cum_results
            else:
                game_results[f"{re.split('_',strat)[1]}_hit_on_{hit_val}"] = cum_results
    
    return game_results
```

```python
def line_chart(conditions_dict, sim_dict): #data_list, sim
    '''
    Create a line chart where data_list is a list of lists of tuples structured (x,y). 
    Each list in data_list is the points for one line on the chart.
    '''
    legend_labels = [key for key,value in sim_dict.items()]
    title = conditions_dict['chart title'] if 'chart title' in conditions_dict.keys() else 'Line Chart'
    x_label = conditions_dict['x-axis title'] if 'x-axis title' in conditions_dict.keys() else 'x-axis'
    y_label = conditions_dict['y-axis title'] if 'y-axis title' in conditions_dict.keys() else 'y-axis'
    style = conditions_dict['chart style'] if 'chart style' in conditions_dict.keys() else 'whitegrid'
    figsize = conditions_dict['figsize'] if 'figsize' in conditions_dict.keys() else (15,6)
    omitFirstx = conditions_dict['omit first x trials'] if 'omit first x trials' in conditions_dict.keys() else 0
    
    sns.set_style(style)
    plt.figure(figsize=figsize)

    for idx,(key,data) in enumerate(sim_dict.items()):
        data = data[omitFirstx:]
        
        # Unzip the list of tuples into two lists, one for x values and one for y values
        x_values, y_values = zip(*data)

        # Create the line chart using seaborn
        line = sns.lineplot(x=x_values, y=y_values, label=(legend_labels[idx] if legend_labels else None))

        # Add a red circle to the last point
        last_point = data[-1]
        plt.scatter(last_point[0], last_point[1], color='red', s=100, edgecolors='black', label=None)

        # Display the y-value of the last point
        plt.annotate(f"{last_point[1]}", (last_point[0], last_point[1]), textcoords="offset points", xytext=(0,10), ha='center')

    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.axhline(y=0, linestyle='--', c='red') #add line to y=0

    # Display win percentages for each dataset
    win_percentage = []
    text_statement = []
    for idx,(key,data) in enumerate(sim_dict.items()):
        last_point = data[-1]
        scl_ratio = float(str((last_point[1] - (-1)) / (1 - (-1))*100)[:4]) #turn win/loss ratio into win %
        win_percentage.append(scl_ratio)
        text_statement.append(f"{legend_labels[idx]}: Based on the results of {last_point[0]+1} games, you can expect to win {scl_ratio}% of the time.")
    sorted_rankings = sorted(zip(win_percentage,text_statement), key=lambda x: x[0], reverse=True)
    for win,txt in sorted_rankings:
        print(txt)
        
    if legend_labels is not None:
        plt.legend()
    plt.show()
```

```python
conditions_dict = {
    'seed':686,
    'trials':10000,
    'number of decks':4,
    'strategies':['strategy_hard', 'strategy_soft'],
    'when to stand':[11,14,17,20],
    'chart title':'Blackjack Cumulative Win/Loss Ratio',
    'y-axis title':'ratio',
    'x-axis title':'number of trials',
    'chart style':'whitegrid',
    'figsize':(15, 10),
    'omit first x trials':50
}
sim_dict = niceWatch_runIt(conditions_dict)
line_chart(conditions_dict, sim_dict)
```
